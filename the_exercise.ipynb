# **Check for Outliers**



# **Use the Interquartile range(IQR) to handle the outliers.**

### **Exercise - Import iqr from seaborn library. This is required in order to calculate the percentile to get the interquartile range**

```python
from seaborn.utils import iqr
```

### **Exercise - Calculate the first quartile(Q1) and third quartile(Q3) of **SepalWidthCm column** using the 25th and 75th percentile.**

Q1 is the first quartile of the data, i.e., to say 25% of the data lies between minimum and Q1.

Q3 is the third quartile of the data, i.e., to say 75% of the data lies between minimum and Q3.

```python
q1 = iris_data['SepalWidthCm'].quantile(0.25)
q3 = iris_data['SepalWidthCm'].quantile(0.75)
```

```python
print(q1)
print(q3)
```

### **Exercise - Calculate the IQR by substracting Q1 from Q3**

The difference between Q3 and Q1 is called the Inter-Quartile Range or IQR.

The difference between the Q3 and Q1 tells us about the range of dataset

```python
exception('print the IQR')
iqr = q3-q1
print()
```

### **Exercise - Calculate the lower and upper bound for the column by multiplying the IQR by 1.5 and subtract it from the value of the Q1 and Q3 respectively.**

The lower and upper bounds are smallest and largest values of the data range. Any values smaller or larger than bounds value are outliers

```python
lower_bound = q1 - (iqr*1.5)
upper_bound = q3 + (iqr*1.5)
print(lower_bound)
print(upper_bound)
```

### **To Do - Check for the shape of the outlier**

```python
outlier = iris_data[(iris_data ['SepalWidthCm'] < lower_bound)|(iris_data['SepalWidthCm'] > upper_bound)]
outlier.shape
```

### **Exercise - Replace the outliers higher than the upper bound with the upper bound and outlier lower than the lower bound with the lower bound**

```python
iris_data['SepalWidthCm'] = np.where(iris_data['SepalWidthCm']> upper_bound, upper_bound, np.where(iris_data['SepalWidthCm'] < lower_bound, lower_bound, iris_data['SepalWidthCm']))
```

# **Confusion matrix**
A confusion matrix is a tabular summary of the number of correct and incorrect predictions made by a classifier.
It can be used to evaluate the performance of a classification model through the calculation of performance metrics like accuracy, precision, recall, and F1-score.
The confusion matrix is a much better way to evaluate the performance of a classifier
Each row represents a class, each column a prediction:



| | Actual | |
| --- |: --- |: --- |
| **Prediction** | True Positive | False Positive |
| | False Negative | True Negative |

[Article on confusion matrix](https://medium.com/analytics-vidhya/what-is-a-confusion-matrix-d1c0f8feda5)

### **Exercise - Find confusion matrix**
*N.B: To do this we have to import confusion_matrix from sklearn.metrics*

```python
# Find confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_test_pred)
cm
```

A Classification report is used to measure the quality of predictions from a classification algorithm. How many predictions are True and how many are False. More specifically, True Positives, False Positives, True negatives and False Negatives are used to predict the metrics of a classification report as shown below.

Precision – What percent of your predictions were correct?

Recall – What percent of the positive cases did you catch?

F1 score – What percent of positive predictions were correct?
